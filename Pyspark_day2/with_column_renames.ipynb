```python
muni=spark.read.table("workspace.munireddy.katigari_det")
display(muni)
```

```python
muni1=muni.withColumn("New_Salary",muni.Salary*5)
muni1.show()
```

```python
muni2=muni1.withColumn("New_Bonus",muni1.New_Salary*0.10)
muni2.show()
```

```python
display(muni2)
```

```python
from pyspark.sql.functions import lit
muni3=muni2.withColumn("Country",lit("India"))
muni3.show()
```

```python
muni4=muni3.withColumn("2026_Increment",lit(0))
muni4.show()
```

```python
muni4.printSchema()
```

```python
from pyspark.sql.functions import col
muni5=muni4.withColumn("New_Bonus",col("New_Bonus").cast("Float"))\
    .withColumn("2026_Increment",col("2026_Increment").cast("Float"))
muni5.printSchema()
display(muni5)
```

```python
from pyspark.sql.functions import lit
m=spark.read.table("workspace.munireddy.katigari_det")
display(m)
```

```python
m1=m.withColumn("Increament",lit(0))
m1.show()
```

```python
m2=m1.withColumn("2026_Increment",lit(10000))
m2.show()
```

```python
from pyspark.sql.functions import col
m3=m2.withColumn("Total_Credited",col("Salary")+col("Increament")+col("2026_Increment"))
m3.show()
```

```python
m3.printSchema()
```

```python
m4=m3.withColumnRenamed("Total_Credited","Final_Amount")
m4.show()
```

```python
m4.write.mode("overwrite").format("delta").saveAsTable("katigari_data")
```

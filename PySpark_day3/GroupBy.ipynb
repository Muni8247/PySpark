```python
n=spark.read.table("workspace.default.katigari_data")
n.show()
```

```python
n1=n.groupBy("Age")
display(n1)
```

```python
n2=n.groupBy("Gender").sum("Salary")
n2.show()
```

```python
from pyspark.sql.functions import lit
n3=n.withColumn("Country",lit("India"))
n3.show()
```

```python
from pyspark.sql.functions import col, when, lit

n3 = n.withColumn("Country", 
    when(col("Name1") == "Swarna", lit("USA"))
    .when(col("Name1") == "Keerthi", lit("UK"))
    .when(col("Name1") == "Muni Reddy", lit("Canada"))
    .when(col("Name1") == "Lakshmi Devi", lit("Canada"))
    .when(col("Name1") == "Naveen", lit("USA"))
    .otherwise(lit("India"))
)

n3.show()
```

```python
n3.write.mode("overwrite").format("delta").saveAsTable("my_family_data")
```

###INNER JOIN

```python
table1=spark.read.table("workspace.default.katigari_data")
table2=spark.read.table("workspace.default.katigari")
```

```python
lpu=table1.join(table2,table1.Id==table2.Id,'inner')
lpu.show()
```

###leftjoin

```python
lpu2=table1.join(table2,table1.Id==table2.Id,'left')
lpu2.show()
```

###RIGHT JOIN

```python
lpu3=table1.join(table2,table1.Id==table2.Id,'right')
lpu3.show()
```

###FULL OUTER

```python
lpu4=table1.join(table2,table1.Id==table2.Id,'full')
lpu4.show()
```

###LEFT SEMI

```python
lpu5=table1.join(table2,table1.Id==table2.Id,'left_semi')
lpu5.show()
```

###LEFT ANTI

```python
lpu6=table1.join(table2,table1.Id==table2.Id,'left_anti')
lpu6.show()
```

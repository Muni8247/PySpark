```python
from pyspark.sql.functions import *

d1=[("Muni","01-09-2000"),("Swarna","21-08-1999")]
df=spark.createDataFrame(d1,["Name","BOD"])
df.show()
                         
```

###date_add()

```python
from pyspark.sql.functions import *
df1 = df.withColumn("1st_Birth_date", date_add(to_date(col("BOD"), "dd-MM-yyyy"), 365))
df1.show()
```

###DATE_SUB

```python
from pyspark.sql.functions import *
df2=(["Engagement","09-08-2026"],["Marriage","10-10-2026"],["First_night","15-10-2026"])
event_dates=spark.createDataFrame(df2,["Event","Date"])
event_dates.show()


df3 = event_dates.withColumn("Reamainder_Dates", date_sub(to_date(col("Date"), "dd-MM-yyyy"), 10))
df3.show()
```

###date_diff

```python
from pyspark.sql.functions import datediff, col, to_date
df4=(["Muni","19-10-2024","10-12-2024"],["Swarna","04-10-2020","08-12-2020"])
event1=spark.createDataFrame(df4,["Name","Enagagement","Marriage"])
event1.show()

df5=event1.withColumn("Days_gap",datediff(to_date(col("Marriage"),"dd-MM-yyyy"),to_date(col("Enagagement"),"dd-MM-yyyy")))
df5.show()
```

###YEAR_DATE

```python
from pyspark.sql.functions import year, to_date, col
df6=(["Muni","01-09-2000","04-04-2017","04-10-2023","10-10-2028"],["Swarna","21-08-1999","04-04-2015","04-04-2019","06-07-2020"])
df7=spark.createDataFrame(df6,["Name","DOB","10th","B.Tech","Marriage"])
df7.show()

f8=(df7
    .withColumn("DOB Year", year(to_date(col("DOB"), "dd-MM-yyyy")))
    .withColumn("10th Year", year(to_date(col("10th"), "dd-MM-yyyy")))
    .withColumn("B.Tech Year", year(to_date(col("`B.Tech`"), "dd-MM-yyyy"))) 
    .withColumn("Marriage Year", year(to_date(col("Marriage"), "dd-MM-yyyy")))
)

f8.show()
```

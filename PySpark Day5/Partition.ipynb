```python
from pyspark.sql.functions import *
df=(["Muni","Male","IT"],["Ramu","Male","Bank"],["Ravi","Male","Bank"],["Rajesh","Male","Bank"],["Keerthi","Female","IT"],["Chandrika","Female","IT"],["Swarna","Female","IT"])
df1=spark.createDataFrame(df,['Name','Gender','Department'])
df1.show()
```

###PARTITION

```python
df1.write.mode("overwrite").partitionBy("Department").format("/Workspace/Users/katigarimuni@gmail.com/PySpark/Data_Frames_Partition")
print("Files successfully created on disk!")
```

```python
df1.write.mode("overwrite").partitionBy("Department").saveAsTable("default.employee_partitions")
```

```python
df=spark.read.table("workspace.default.employee_partitions")
df.show()
```

```python
df1.write.mode("overwrite").partitionBy("Gender").saveAsTable("default.employee_partitions_Gender")
```

```python
df2=spark.read.table("workspace.default.employee_partitions_gender")
df2.show()
```

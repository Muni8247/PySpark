```python
from pyspark.sql import SparkSession
from pyspark.sql.types import StructType, StructField, StringType, IntegerType, LongType

# Initialize Spark Session
spark = SparkSession.builder.appName("SchemaExample").getOrCreate()

# 1. Define a sub-schema (Address Details)
address_schema = StructType([
    StructField("City", StringType(), True),
    StructField("State", StringType(), True),
    StructField("District", StringType(), True)
])

# 2. Define the Main Schema (Nesting the address_schema inside it)
employee_schema = StructType([
    StructField("Id", IntegerType(), False),          # Cannot be null
    StructField("Name", StringType(), False),         # Cannot be null
    StructField("Salary", LongType(), True),
    
    # This column holds a structural group of 3 fields inside it!
    StructField("LocationDetails", address_schema, True) 
])

# 3. Create raw data matching this exact nested blueprint
raw_data = [
    (1, "Muni Reddy", 74000, ("Bangalore", "Karnataka", "Bangalore Urban")),
    (2, "Swarna", 38000, ("Kadapa", "Andhra Pradesh", "YSR Kadapa")),
    (3, "Parthu", 74000, ("Chennai", "Tamil Nadu", "Chennai"))
]

# 4. Create the DataFrame using your custom schema
df = spark.createDataFrame(raw_data, schema=employee_schema)
df.show()
```

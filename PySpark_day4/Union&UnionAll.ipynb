```python
%sql
insert into n2 values
(1,"Ramu",33),
(2,"Chandrika",31),
(3,"Yakshith",16),
(4,"Yash",4);
```

```python
lpu=spark.read.table("workspace.default.n1")
lpu1=spark.read.table("workspace.default.n2")
```

```python
lpu.printSchema()
lpu1.printSchema()
```

###UNION & UNIONALL

```python
merge=lpu.union(lpu1)
merge.show()
```

###UNIONALL

```python
merge1=lpu.unionAll(lpu1)
merge1.show()
```

```python
import pyspark
from pyspark.sql import SparkSession

n3=[("Gopal",56,1),\
    ("Rushi",51,2),\
    ("Muni D",40,3),\
    ("Raidu",42,4)\
]
columns=["Name","Age","Id"]
n4=spark.createDataFrame(data=n3,schema=columns)
n4.show()
```

###UNION BY NAME

```python
n5=merge1.unionByName(n4)
n5.show()
```

###DISTINCT UNION

```python
n6=merge1.union(n5).distinct()
n6.show()
```
